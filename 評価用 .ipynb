Word Net

```python
from nltk.corpus import wordnet as wn

def get_synonyms(word):
    synonyms = set()

    for word in wn.synsets(word)[:5]:
        synonyms.add(str(word.name()).split('.')[0])

    return(synonyms)

```

Transformer Pipeline

```python
from transformers import pipeline
model = pipeline("fill-mask")

def fill_mask(word, num, text): #word <- 空欄にする単語     num <- wordの対象が複数ある時に対象とする場所を示す     text <- sentence
    doc = nlp(text)
    mask_text = ""
    count = 0
    for token in doc:
        if token.text == word:
            if num == count:
                mask_text += "{} "
            else:
                mask_text += token.text + " "

            count += 1   
        else:
            mask_text += token.text + " "
    #print(mask_text)
    
    inputs = [mask_text.format(model.tokenizer.mask_token)]
    #print(inputs)
    res = model(inputs, top_k=5)
    return [str(r["token_str"]).replace(" ", "") for r in res]

fill_mask("watch", 0, "I turn on the tv and sit in the sofa. I watch the tv.")
```

Convert to Present tense

```python
def present_tense(text): # 動詞の活用形を現在形に直して返す
    for token in nlp(str(text).replace("_"," ")):
        if token.pos_== 'VERB':
            text = str(text).replace(token.text, token.lemma_)
    return text
```

```python
def decision_action_object(verb, pharasal_verb, object, sentence, num):

    decision_action = None
    decision_object = []

    # 動詞の決定
    if pharasal_verb:
        if present_tense(pharasal_verb) in action_list: # 句動詞がアクションリストに存在するか
            decision_action = str(present_tense(pharasal_verb))

        elif get_synonyms(present_tense(pharasal_verb)) & set(action_list): # 句動詞の同義語（WordNetで検索）がアクションリストに存在するか
            mached_list = list(get_synonyms(present_tense(pharasal_verb)) & set(action_list))

            if len(mached_list)==1:
                decision_action = mached_list[0]
            #else:
                #類似度の高いほう

    if decision_action is None:
        if present_tense(verb) in action_list: # 動詞がアクションリストに存在するか
            decision_action = present_tense(verb)

        elif get_synonyms(present_tense(verb)) & set(action_list): # 動詞の同義語（WordNetで検索）がアクションリストに存在するか
            mached_list = list(get_synonyms(present_tense(verb)) & set(action_list))

            if len(mached_list)==1:
                decision_action = mached_list[0]
            #else:
                #類似度の高いほう

        else:
            candidate_action = fill_mask(verb, num, sentence) # 文章中の動詞を空欄にして空欄部分に入りそうな動詞の候補を取得
            #print(candidate_action)
            for action in candidate_action: 
                action = present_tense(action)
                if action in action_list: # 候補中の動詞がアクションリストに存在すれば決定する
                    decision_action = action
                    break
                elif pharasal_verb: # 句動詞がある場合 候補中の動詞+前置詞 がアクションリストに存在するか
                    action = action + "_" + str(pharasal_verb).rsplit("_")[1]
                    if action in action_list:
                        decision_action = action
                        break
                #<- 複数存在する場合にどうするか？　オブジェクトのプロパティから実行できるか？　部屋の場合はWalkにしてしまうのか？

            #if decision_action is None: # 見つからなかったときに類似度で出すか？
                #decision_action = get_similar_action(verb)

    for obj in object:

        if obj in object_list: # オブジェクトがオブジェクトリストに存在するか
            decision_object.append(obj)

        elif obj in equivalence_object: # オブジェクトがVirtualhomeが定義しているオブジェクトリスト中の同義語一覧に存在する場合
            decision_object.append(list(set(object_name_equivalence.get(obj)) & set(object_list))[0]) # オブジェクトのVirtualhomeで定義された同義語とオブジェクトリストの共通集合から取得

        # "_" が文字列に含まれている場合
        elif "_" in str(obj): 
            obj = str(obj).replace("_","")  # "_" を外して上記と同じことをする
            if obj in object_list:
                decision_object.append(obj)

            elif obj in equivalence_object:
                decision_object.append(list(set(object_name_equivalence.get(obj)) & set(object_list))[0])

        elif set(get_synonyms(obj)) & set(object_list):
            mached_list = list(get_synonyms(obj) & set(object_list))

            if len(mached_list)==1:
                decision_object.append(mached_list[0])

            #else:
                #類似度の高いほう

        elif set(get_synonyms(obj)) & set(equivalence_object):
            mached_list = list(get_synonyms(obj) & set(equivalence_object))
            #print(mached_list)

            if len(mached_list)==1:
                decision_object.append(list(set(object_name_equivalence.get(mached_list[0])) & set(object_list))[0])

            #else:
                #類似度の高いほう

        elif get_similar_object(obj):
            decision_object.append(get_similar_object(obj))
    
    if "" in decision_object:
        decision_object.remove("")
    
    return decision_action, decision_object
    
```

Morphological analysis

```python
def morphological_analysis(text):
    sentence_list = []
    sentence = ""

    for i, token in enumerate(text):

        if token.text==".":
            sentence = sentence.rstrip() + "."
            sentence_list.append(nlp(sentence))
            sentence = ""
        elif len(text)-1 == i:
            sentence += str(token.text).lower() + "."
            sentence_list.append(nlp(sentence))
        elif sentence == "" and token.tag_ != "PRP":#token.pos_ == "VERB"
            sentence += "I " + str(token.text).lower() + " "
        elif token.tag_ == "PRP":
            sentence += token.text + " "
        else:
            sentence += str(token.text).lower() + " "    

    #print(sentence_list)
    doc = ""
    for sentence in sentence_list:
        doc += str(sentence) + " "

    pair_dict_list = []
    for subActivity_doc in sentence_list:
        verb = None     #動詞
        prep = None     #前置詞修飾詞
        prt = None      #句動詞助詞
        comp = None     #補語
        object = []     #目的語
        compound = None #複合修飾詞
        pharasal_verb = None    #句動詞
        #adverb = None  #副詞

        for token in subActivity_doc:
            #print(token.text, token.tag_, token.pos_, token.head.text, token.dep_)

            if (token.dep_ == "ROOT" or token.dep_ == "conj" or token.dep_ == "dep" or token.dep_ == "advcl" or token.dep_ == "ccomp" or token.dep_ == "xcomp" or token.dep_ == "pcomp") and token.pos_ == "VERB": #token.pos_ == "VERB": 
                if verb or pharasal_verb:
                    # 辞書形式でアクションとオブジェクトのペアを追加
                    pair_dict = {}
                    if pharasal_verb:
                        pair_dict.setdefault("action", pharasal_verb)
                        pharasal_verb = None
                    else:
                        pair_dict.setdefault("action", verb)

                    object =  [obj for obj in object]
                    pair_dict.setdefault("object", object)
                    #pair_dict.setdefault("sentence", doc)
                    pair_dict_list.append(pair_dict)
                    # 空にする
                    verb = None
                    object = []

                verb = token.text

            elif token.dep_ == "dobj" and token.head.text == verb or token.dep_ == "pobj" and (token.head.text == prep or token.head.text == prt):
                if token.text == "it":
                    if pair_dict_list:
                        if pair_dict_list[-1].get("object"):
                            object.append(pair_dict_list[-1].get("object")[-1])
                elif compound:#<- 
                    object.append(compound + "_" + token.text)
                    compound = None
                else:
                    object.append(token.text)

            elif token.dep_ == "prt" and token.head.text == verb:
                prt = token.text
                if not pharasal_verb:
                    pharasal_verb = verb + "_" + prt

            elif token.dep_ == "prep" and token.head.text == verb:
                prep = token.text
                if not pharasal_verb:
                    pharasal_verb = verb + "_" + prep #改善
            
            elif token.dep_ == "conj" and token.pos_ == "NOUN":
                if compound:#<- 
                    object .append(compound + "_" + token.text)
                    compound = None
                else:
                    object.append(token.text)

            elif token.dep_ == "compound":
                compound = token.text

            elif token.dep_ == "punct" and token.text == ".":
                # 辞書形式でアクションとオブジェクトのペアを追加
                pair_dict = {}
                if pharasal_verb:
                    pair_dict.setdefault("action", pharasal_verb)
                    pharasal_verb = None
                else:
                    pair_dict.setdefault("action", verb)

                object =  [obj for obj in object]   
                pair_dict.setdefault("object", object)
                #pair_dict.setdefault("sentence", doc)
                pair_dict_list.append(pair_dict)

    #print(pair_dict_list)
    #print(doc)
    return pair_dict_list, doc
```

```python
def reset_action(action_script):
    script = []
    for action in action_script:
        action[0] = str(action[0]).replace("_","")
        if len(action)==1:
            script.append(f"<char0> [{action[0]}]")
        elif len(action)==2:
            script.append(f"<char0> [{action[0]}] <{action[1]}> (1)")
        elif len(action)==3:
            script.append(f"<char0> [{action[0]}] <{action[1]}> (1) <{action[2]}> (1)")
    
    comm.reset(scene_num-1)
    comm.add_character('Chars/male1', initial_room='bathroom')
    success, graph = comm.environment_graph()
    comm.render_script(script=script, find_solution=True, skip_execution=False, recording=True, camera_mode=['AUTO'])
```

```python
def execution_confirmation(s):
    print(s)
    s = str(s).replace("_","")
    script = []
    script.append(s)
    success, graph = comm.environment_graph()
    a, b = comm.render_script(script=script, find_solution=True, skip_execution=False, recording=True, camera_mode=['AUTO'])
    return a
```

```python
import math
def calc_distance(a, b):
    x = abs(a[0] - b[0])
    z = abs(a[2] - b[2])
    y = math.sqrt(x*x + z*z)
    
    return y
```

```python
def search_nearest_object(object_name):
    success, graph = comm.environment_graph()
    character = [n for n in graph["nodes"] if n.get("class_name")=="character"][0]
    chara_position = character.get("obj_transform").get("position")
    
    objects = [n for n in graph["nodes"] if n.get("class_name") == object_name]

    nearest_object_id = 0
    distance = 100
    
    for o in objects:
        object_position = o.get("obj_transform").get("position")
        if calc_distance(chara_position, object_position) < distance:
            distance = calc_distance(chara_position, object_position)
            nearest_object_id = o.get("id")
    
    return nearest_object_id, distance
```

```python
def add_standup(action_script):
    global SITTING
    
    if SITTING == True:
        if execution_confirmation('<char0> [standup]') == True:
            action_script.append(["stand_up"])
            SITTING = False
        else:
            reset_action(action_script)
            
    return action_script
```

```python
def add_walk_run(action_script, action, object, distance):
    
    if action_script and object in room_list:
    
        if action_script[-1][0] == "walk" or action_script[-1][0] == "run":
            
            if action_script[-1][1] in room_list:
                action_script = action_script[:-1]
                reset_action(action_script)
    
    if distance > 1:
        
        if execution_confirmation(f'<char0> [{action}] <{object}> (1)') == True:
            action_script.append([action, object])
        else:
            reset_action(action_script)
    
    return action_script
```

```python
def add_find(action_script, object):
    if object not in find_object:
        
        if execution_confirmation(f'<char0> [find] <{object}> (1)') == True:
            action_script.append(["find", object])
            find_object.append(object)
        else:
            reset_action(action_script)
        
    return action_script        
```

```python
def add_grab(action_script, object):
    
    success, graph = comm.environment_graph()
    object_id, distance = search_nearest_object(object)
    properties = [n.get("properties") for n in graph["nodes"] if n.get("id") == object_id][0]
    
    if ("GRABBABLE" in properties) and (object not in grab_object) and ( len(grab_object) < 2 ):
        
        if execution_confirmation(f'<char0> [grab] <{object}> (1)') == True:
            action_script.append(["grab", object])
            grab_object.append(object)
        else:
            reset_action(action_script)
    
    return action_script
```

```python
def add_open(action_script, object):
    
    success, graph = comm.environment_graph()
    object_id, distance = search_nearest_object(object)
    properties = [n.get("properties") for n in graph["nodes"] if n.get("id") == object_id][0]
    states = [n.get("states") for n in graph["nodes"] if n.get("id") == object_id][0]
    
    if ("CAN_OPEN" in properties) and ("CLOSED" in states) and ( len(grab_object) < 2 ):
        
        if execution_confirmation(f'<char0> [open] <{object}> (1)') == True:
            action_script.append(["open", object])
        else:
            reset_action(action_script)
            
    return action_script
```

Get Current Location

```python
def current_location():
    success, graph = comm.environment_graph()
    character = [n for n in graph["nodes"] if n.get("class_name")=="character"][0]
    
    room_id = [e.get("to_id") for e in graph["edges"] if character.get("id") == e.get("from_id") and e.get("relation_type") == "INSIDE"][0]
    room_name = ""
    
    if room_id == bathroom_id:
        room_name = "bathroom"
    
    elif room_id == bedroom_id:
        room_name = "bedroom"
    
    elif room_id == kitchen_id:
        room_name = "kitchen"
        
    elif room_id == livingroom_id:
        room_name = "livingroom"
    
    return room_name
```

Add Action

```python
def add_action(action_script, action, object):
    success, graph = comm.environment_graph()
    
    global SITTING
    object_id = 0
    distance = 100
    properties = []
    states = []
    
    if object:
        object_id, distance = search_nearest_object(object[0])
        properties = [n.get("properties") for n in graph["nodes"] if n.get("id") == object_id][0]
        states = [n.get("states") for n in graph["nodes"] if n.get("id") == object_id][0] 
    
    if action in noObjAction:
        
        if action == "stand_up":
            action_script = add_standup(action_script)
            
        else:
            action_script = add_standup(action_script)
            
            if execution_confirmation(f'<char0> [{action}]') == True:
                action_script.append([action])
            else:
                reset_action(action_script)
            
    
    elif action=="walk" or action=="run":
        action_script = add_standup(action_script)
        action_script = add_walk_run(action_script, action, object[0], distance)

        
    elif action=="find":
        
        if distance > 1:
            action_script = add_standup(action_script)
            action_script = add_walk_run(action_script, "walk", object[0], distance)
            
        action_script = add_find(action_script, object[0])
    
    
    elif action == "watch":
        
        objects_id = [n.get("id") for n in graph["nodes"] if n.get("class_name") == object[0]]
        room_name = ""
            
        for object_id in objects_id:  
            room_id = [e.get("to_id") for e in graph["edges"] if object_id == e.get("from_id") and e.get("relation_type") == "INSIDE"][0]
            if room_id == bathroom_id and current_location() == "bathroom":
                room_name = "bathroom"
                break
    
            elif room_id == bedroom_id and current_location() == "bedroom":
                room_name = "bedroom"
                break
    
            elif room_id == kitchen_id and current_location() == "kitchen":
                room_name = "kitchen"
                break
        
            elif room_id == livingroom_id and current_location() == "livingroom":
                room_name = "livingroom"
                break
                    
        if "HAS_SWITCH" in properties:
                
            if "OFF" in states:
                action_script = add_standup(action_script)
                action_script = add_walk_run(action_script, "walk", object[0], distance)
                action_script = add_find(action_script, object[0])
                    
                if execution_confirmation(f'<char0> [switchon] <{object[0]}> (1)') == True:
                    action_script.append(["switchon", object[0]])
                else:
                    reset_action(action_script)
            
            elif room_name == "":
                action_script = add_standup(action_script)
                action_script = add_walk_run(action_script, "walk", object[0], distance)
                action_script = add_find(action_script, object[0])
                
            else:
                action_script = add_find(action_script, object[0])
                
        else:
            
            if room_name == "":
                action_script = add_standup(action_script)
                action_script = add_walk_run(action_script, "walk", object[0], distance)
                action_script = add_find(action_script, object[0])
                
            else:
                action_script = add_find(action_script, object[0])
                
                
        if execution_confirmation(f'<char0> [watch] <{object[0]}> (1)') == True:
            action_script.append(["watch", object[0]])
        else:
            reset_action(action_script)
            
            
    else:
        action_script = add_standup(action_script)
        action_script = add_walk_run(action_script, "walk", object[0], distance)
        action_script = add_find(action_script, object[0])
    
        if action == "sit":
            
            if "SITTABLE" in properties:
                if execution_confirmation(f'<char0> [sit] <{object[0]}> (1)') == True:
                    action_script.append(["sit", object[0]])
                    SITTING = True
                else:
                    reset_action(action_script)
            
            
        elif action == "grab":
            action_script = add_grab(action_script, object[0])

                
        elif action == "open":
            action_script = add_open(action_script, object[0])

                    
        elif action == "close":
    
            if ("CAN_OPEN" in properties) and ("OPEN" in states) and ( len(grab_object) < 2 ):
        
                if execution_confirmation(f'<char0> [close] <{object[0]}> (1)') == True:
                    action_script.append(["close", object[0]])
                else:
                    reset_action(action_script)

                    
        elif action == "put_on" and len(object) > 1:
            
            if len(grab_object) < 2:
                action_script = add_grab(action_script, object[0])
                
                object_id, distance = search_nearest_object(object[1])
                
                action_script = add_walk_run(action_script, "walk", object[1], distance)
                action_script = add_find(action_script, object[1])
                
                # VirtualHome での表記上，putback > puton となる．
                if execution_confirmation(f'<char0> [putback] <{object[0]}> (1) <{object[1]}> (1)') == True:
                    action_script.append(["putback", object[0], object[1]])
                    grab_object.remove(object[0])
                else:
                    reset_action(action_script)

            
        elif action == "put_in" and len(object) > 1:
            
            if len(grab_object) == 0:
                action_script = add_grab(action_script, object[0])
                
                object_id, distance = search_nearest_object(object[1])
                properties = [n.get("properties") for n in graph["nodes"] if n.get("id") == object_id][0]
        
                action_script = add_walk_run(action_script, "walk", object[1], distance)
                action_script = add_find(action_script, object[1])
                action_script = add_open(action_script, object[1])
        
                if "CONTAINERS" in properties:
        
                    if execution_confirmation(f'<char0> [putin] <{object[0]}> (1) <{object[1]}> (1)') == True:
                        action_script.append(["putin", object[0], object[1]])
                        grab_object.remove(object[0])
                    else:
                        reset_action(action_script)      
                

        elif action == "switch_on":
            
            if "HAS_SWITCH" in properties and  "OFF" in states:
                if execution_confirmation(f'<char0> [switchon] <{object[0]}> (1)') == True:
                    action_script.append(["switchon", object[0]])
                else:
                    reset_action(action_script)

                
        elif action == "switch_off":
            
            if "HAS_SWITCH" in properties and  "ON" in states:
                if execution_confirmation(f'<char0> [switchoff] <{object[0]}> (1)') == True:
                    action_script.append(["switchoff", object[0]])
                else:
                    reset_action(action_script)

                
        elif action == "drink":
            
            if "RECIPIENT" in properties and "POURABLE" in properties:
                
                if object[0] not in grab_object:
                    action_script = add_grab(action_script, object[0])
                    
                if execution_confirmation(f'<char0> [drink] <{object[0]}> (1)') == True:
                    action_script.append(["drink", object[0]])
                else:
                    reset_action(action_script)

                    
        elif action == "look_at":
            
            if execution_confirmation(f'<char0> [turnto] <{object[0]}> (1)') == True:
                action_script.append(["turnto", object[0]])
            else:
                reset_action(action_script)
                
            if execution_confirmation(f'<char0> [lookat] <{object[0]}> (1)') == True:
                action_script.append(["lookat", object[0]])
            else:
                reset_action(action_script)
            

        elif action == "read":
            
            if "READABLE" in properties:
                
                if object[0] not in grab_object:
                    action_script = add_grab(action_script, object[0])
                    
                if execution_confirmation(f'<char0> [read] <{object[0]}> (1)') == True:
                    action_script.append(["read", object[0]])
                else:
                    reset_action(action_script)
    
    return action_script
```

Generate Script

```python
def generate_script(text):

    action_script = []

    #依存構造解析の情報を保存 
    row = morphological_analysis(nlp(text))
    text_info = row[0]
    sentence = row[1]
    verb_list = [str(info.get("action")).rsplit("_")[0] for info in text_info] # 動詞の一覧

    for num, info in enumerate(text_info):
        if "_" in str(info.get("action")): # 動詞+前置詞の場合 動詞と句動詞2パターンに分解
            verb = str(info.get("action")).rsplit("_")[0]
            pharasal_verb = str(info.get("action"))
        else:
            verb = str(info.get("action"))
            pharasal_verb = None
        
        if verb_list[:num+1].count("verb") > 1: # 同じ動詞が存在する場合に前から何番目か計算 <- fill_mask で空欄にする場所を知っておきたい
            row = decision_action_object(verb, pharasal_verb, info.get("object"), sentence, verb_list[:num+1].count("verb")-1)
        else:
            row = decision_action_object(verb, pharasal_verb, info.get("object"), sentence, 0)
        action = row[0]
        object = row[1]
        #print(action, object)

        # アクションとオブジェクト両方ある場合
        if action and object:
            
            # オブジェクトがある部屋をカウント
            count = 0
            for value in room_dict.values():
                if object[0] in value:
                    count += 1
                    
            # オブジェクトが部屋の場合
            if object[0] in room_list:
                if action=="walk" or action=="run" and current_location() != object[0]:
                    action_script = add_action(action_script, action, object)
                
            # オブジェクトがある部屋が複数ある
            elif count>1:
                
                object_in_current_room = False
                
                for key, value in room_dict.items():
                    # 現在いる部屋にオブジェクトが存在する場合
                    if object[0] in value and current_location() == key:
                        action_script = add_action(action_script, action, object)
                        object_in_current_room = True
                        break

                # 現在いる部屋にオブジェクトが存在しなかった場合、近いほうのオブジェクトを選択し、その部屋まで移動
                if object_in_current_room == False:
                    
                    nearest_object_id, distance = search_nearest_object(object[0])
                        
                    room_id = [e.get("to_id") for e in graph["edges"] if nearest_object_id == e.get("from_id") and e.get("relation_type") == "INSIDE"][0]
                    if room_id == bathroom_id:
                        action_script = add_action(action_script, "walk", ["bathroom"])
    
                    elif room_id == bedroom_id:
                        action_script = add_action(action_script, "walk", ["bedroom"])
    
                    elif room_id == kitchen_id:
                        action_script = add_action(action_script, "walk", ["kitchen"])
        
                    elif room_id == livingroom_id:
                        action_script = add_action(action_script, "walk", ["livingroom"])
                    
                    action_script = add_action(action_script, action, object)
                        
                    #または、関係なしにアクションスクリプトに追加
                                    
            else:
                #オブジェクトがバスルームにある
                if object[0] in bathroom_objects and current_location() != "bathroom":
                    action_script = add_action(action_script, "walk", ["bathroom"])

                #オブジェクトがベッドルームにある
                elif object[0] in bedroom_objects and current_location() != "bedroom":
                    action_script = add_action(action_script, "walk", ["bedroom"])

                #オブジェクトがダイニングルームにある
                elif object[0] in kitchen_objects and current_location() != "kitchen":
                    action_script = add_action(action_script, "walk", ["kitchen"])

                #オブジェクトがホームオフィスにある
                elif object[0] in livingroom_objects and current_location() != "livingroom":
                    action_script = add_action(action_script, "walk", ["livingroom"])

                action_script = add_action(action_script, action, object)
                
        
        # 引数を必要としないアクションの場合
        elif action in noObjAction:
            action_script = add_action(action_script, action, object)
              
    return action_script
```

```python
import glob

files = glob.glob(f"dataset/{scene_num}/executable/*.txt")

accuracy_list = []
recall_list = []

action_accuracy_list = []
action_recall_list = []

object_accuracy_list = []
object_recall_list = []

error_files = []

mistake_objects = []

for file in files:
    comm.reset(scene_num-1)
    comm.add_character('Chars/male1', initial_room='bathroom')
    success, graph = comm.environment_graph()

    global SITTING, find_object, grab_object
    SITTING = False
    find_object = []
    grab_object = [] 
    print(file)

    with open(file, 'r') as f:
        text = f.readlines()[1].strip("\n")
        print("入力文：" + text)

    with open(file, 'r') as f:
        correct_data = []
        correct_script = []
        file_data = f.readlines()[4:]

        for line in file_data:
            if len(line.split()) == 1:
                line_data = []
                line_data.append(str(line).strip("\n").replace("[", "").replace("]", "").lower())
                correct_script.append(line_data)
            elif len(line.split()) == 3:
                line_data = []
                line_data.append(str(line.split()[0]).replace("[", "").replace("]", "").lower())
                line_data.append(str(line.split()[1]).replace("<", "").replace(">", ""))
                correct_script.append(line_data)
            else:
                line_data = []
                line_data.append(str(line.split()[0]).replace("[", "").replace("]", "").lower())
                line_data.append(str(line.split()[1]).replace("<", "").replace(">", ""))
                line_data.append(str(line.split()[3]).replace("<", "").replace(">", ""))
                correct_script.append(line_data)

    try:
        script = generate_script(text)

        # データセットのオブジェクトの表記を変換
        generate_objects = set()
        for s in script:
            if len(s)>1:
                generate_objects.add(s[1])
            elif len(s)>2:
                generate_objects.add(s[1])
                generate_objects.add(s[2])
        generate_objects = list(generate_objects)

        for row in correct_script:
            if len(row)>1:

                if row[1] not in object_list and row[1] in object_name_equivalence:

                    if set(generate_objects) & set(object_name_equivalence.get(row[1])):
                        row[1] = list((set(generate_objects) & set(object_name_equivalence.get(row[1]))))[0]
                    else:
                        row[1] = object_name_equivalence.get(row[1])[0]

            elif len(s)>2:

                if row[1] not in object_list and row[1] in object_name_equivalence:

                    if set(generate_objects) & set(object_name_equivalence.get(row[1])):
                        row[1] = list((set(generate_objects) & set(object_name_equivalence.get(row[1]))))[0]
                    else:
                        row[1] = object_name_equivalence.get(row[1])[0]

                if row[2] not in generate_objects and row[2] in object_name_equivalence:

                    if set(generate_objects) & set(object_name_equivalence.get(row[2])):
                        row[2] = list((set(generate_objects) & set(object_name_equivalence.get(row[2]))))[0]
                    else:
                        row[2] = object_name_equivalence.get(row[2])[0]

        print("出力アクション")
        print(script)
        print("正解アクション")
        print(correct_script)


        #システム出力アクション数
        output_action_num = len(script)
        #システム出力のオブジェクトの数
        output_object_num = 0
        #正解アクション数
        correct_num = len(correct_script)
        #正解オブジェクトの数
        correct_object_num = 0
        for row in correct_script:
            if len(row) > 1:
                correct_object_num += 1
        #システム出力中の正解アクション数
        output_correct_num = 0
        output_action_correct_num = 0
        output_object_correct_num = 0

        
        # 最後に正解した正解アクションスクリプト中の正解アクションの位置
        last_correct_index = -1

        # 出力アクションをリストの前から見ていき，正解アクションに照合する．
        # 正解時は，last_correct_index を更新する．
        # 照合するときは，last_correct_index の次の要素から照らし合わせる．つまり，順序を考慮したいため，正解したアクションより前のアクションは照合しない．

        # ここでは，アクションとオブジェクトのセットで照合
        for row in script:
            
            #次照合する位置
            index = last_correct_index + 1

            if len(row) == 1:

                # 出力アクションと最後に正解したアクション以降のアクションと順に照合
                for i in range(index, len(correct_script)):

                    # 同じ要素数を持っていなければ不正解．
                    if len(row) == len(correct_script[i]):
                        
                        if row[0] == correct_script[i][0]:
                            output_correct_num += 1
                            last_correct_index = i
                            break

            # オブジェクト１つのアクション
            elif len(row) == 2:
                
                for i in range(index, len(correct_script)):

                    if len(row) == len(correct_script[i]):
                        
                        if row[0] == correct_script[i][0] and row[1] == correct_script[i][1]:
                            output_correct_num += 1
                            last_correct_index = i
                            break

            # オブジェクトが２つのアクション
            else:

                for i in range(index, len(correct_script)):

                    if len(row) == len(correct_script[i]):
                        
                        if row[0] == correct_script[i][0] and row[1] == correct_script[i][1] and row[2] == correct_script[i][2]:
                            output_correct_num += 1
                            last_correct_index = i
                            break

            #正解アクションスクリプトの最後の要素に達した場合 
            if len(correct_script)-1 == last_correct_index:
                break


        # アクション（動作）のみの照合
        last_correct_index = -1
        for row in script:

            index = last_correct_index + 1
            
            for i in range(index, len(correct_script)):

                if row[0] == correct_script[i][0]:
                    output_action_correct_num += 1
                    last_correct_index = i
                    break
 
            if len(correct_script)-1 == last_correct_index:
                break


        # オブジェクトのみの照合，オブジェクト（引数）が２つのアクションの場合は，２つ正解の場合に正解とみなす．
        last_correct_index = -1

        for row in script:

            index = last_correct_index + 1

            if len(row) > 1:
                output_object_num += 1

                if len(row) == 2:

                    for i in range(index, len(correct_script)):
                            
                        if len(row) == len(correct_script[i]):

                            if row[1] == correct_script[i][1]:
                                output_object_correct_num += 1
                                last_correct_index = i
                                break

                else:

                    for i in range(index, len(correct_script)):
                            
                        if len(row) == len(correct_script[i]):

                            if row[1] == correct_script[i][1] and row[2] == correct_script[i][2]:
                                output_object_correct_num += 1
                                last_correct_index = i
                                break

            if len(correct_script)-1 == last_correct_index:
                break

        
        #適合率
        accuracy = output_correct_num/output_action_num
        accuracy_list.append(accuracy)

        action_accuracy = output_action_correct_num/output_action_num
        action_accuracy_list.append(action_accuracy)

        object_accuracy = output_object_correct_num/output_object_num
        object_accuracy_list.append(object_accuracy)

        #再現率
        recall = output_correct_num/correct_num
        recall_list.append(recall)

        action_recall = output_action_correct_num/correct_num
        action_recall_list.append(action_recall)

        object_recall = output_object_correct_num/correct_object_num
        object_recall_list.append(object_recall)

        print("適合率：" + "{:.3f}".format(accuracy))
        print("再現率：" + "{:.3f}".format(recall))
        print("適合率（アクション）：" + "{:.3f}".format(action_accuracy))
        print("再現率（アクション）：" + "{:.3f}".format(action_recall))
        print("適合率（オブジェクト）：" + "{:.3f}".format(object_accuracy))
        print("再現率（オブジェクト）：" + "{:.3f}".format(object_recall))
    
    
    except:
        print("Error")
        error_files.append(file)
        pass


print("平均　適合率：" + "{:.3f}".format(sum(accuracy_list)/len(accuracy_list)))
print("平均　再現率：" + "{:.3f}".format(sum(recall_list)/len(recall_list)))
print("平均　適合率（アクション）：" + "{:.3f}".format(sum(action_accuracy_list)/len(action_accuracy_list)))
print("平均　再現率（アクション）：" + "{:.3f}".format(sum(action_recall_list)/len(action_recall_list)))
print("平均　適合率（オブジェクト）：" + "{:.3f}".format(sum(object_accuracy_list)/len(object_accuracy_list)))
print("平均　再現率（オブジェクト）：" + "{:.3f}".format(sum(object_recall_list)/len(object_recall_list)))

def get_variance(l, ave):
    total = 0
    for n in l:
        total += (n-ave)*(n-ave)
    return total/len(l)

print("分散　適合率：" + "{:.3f}".format(get_variance(accuracy_list, sum(accuracy_list)/len(accuracy_list))))
print("分散　再現率：" + "{:.3f}".format(get_variance(recall_list, sum(recall_list)/len(recall_list))))
print("分散　適合率（アクション）：" + "{:.3f}".format(get_variance(action_accuracy_list, sum(action_accuracy_list)/len(action_accuracy_list))))
print("分散　再現率（アクション）：" + "{:.3f}".format(get_variance(action_recall_list, sum(action_recall_list)/len(action_recall_list))))
print("分散　適合率（オブジェクト）：" + "{:.3f}".format(get_variance(object_accuracy_list, sum(object_accuracy_list)/len(object_accuracy_list))))
print("分散　再現率（オブジェクト）：" + "{:.3f}".format(get_variance(object_recall_list, sum(object_recall_list)/len(object_recall_list))))

print("error files: " + str(error_files))
print("number of errors: " + str(len(error_files)))
```
